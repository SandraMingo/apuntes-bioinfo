# Decision Theory

This notebook illustrates the use of class posterior probabilities to make optimal class label assignments using decision theory. We also explore here the use of the reject option and the consideration of different classification error costs, as indicated by a loss or cost matrix. For illustration purposes, we will assume that we have direct access to the class conditional densities $p(\mathbf{x}|\mathcal{C}_k)$, $k=1,\ldots,K$.

First we load the required packages.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import scipy as sp
```

## Class Conditional Densities

The class conditional densities, i.e., $p(\mathcal{x}|\mathcal{C}_k)$ we will be assumed to be given. In this case, we will consider Gaussian densities with some pre-defined parameters (chosen at random). The dimensionality of the problem will be restricted for the sake of visualization. That is, $D = 2$. The number of class labels will be three.

```python
# We fix the random seed to guarantee similar results each time the notebook is run

np.random.seed(0)

D = 2
K = 3
means = [ np.random.normal(size = D) for i in range(K) ]
covariances = [ np.random.normal(size = ((D, D))) for i in range(K) ]

# We are careful to guarantee that the covariance matrices are positive-definite

for i in range(K):
    covariances[ i ] = np.dot(covariances[ i ], covariances[ i ].T)
```

We plot the corresponding Gaussian densities, genrated randomly.

```python
x_min = -2
x_max = 5
y_min = -5
y_max = 5

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

f, axarr = plt.subplots(1, 3, sharex='col', sharey='row', figsize=(15, 5))

for i in range(K):
    mv = stats.multivariate_normal(mean = means[ i ], cov = covariances[ i ])
    Z = mv.pdf(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    axarr[i].set_title('k = ' + str(i))
    axarr[i].set_xlabel('Attribute #1')
    axarr[i].set_ylabel('Attribute #2')
    axarr[ i ].contourf(xx, yy, Z, cmap=plt.cm.hot)
plt.show()
```

## Prior Class Probabilities

We specify here the prior probability distribution for each class. We will assume that the first and third classes are more probable, a priori.

```python
class_prior = np.array([ 0.5, 0.2, 0.3])
```

## Posterior Class Probabilities 

We use Bayes' rule to compute the class posterior probabilities.

```python
def post_prob(x, k):
    
    p_x_and_ck = np.ones((x.shape[ 0 ], K))
    
    for j in range(K):
        mv = sp.stats.multivariate_normal(mean = means[ j ], cov = covariances[ j ])
        p_x_given_ck = mv.pdf(x)
        p_x_and_ck[ :, j ] = p_x_given_ck * class_prior[ j ]
        
    return p_x_and_ck[ :, k ] / np.sum(p_x_and_ck, axis = 1)
    
```

We plot the class posterior probabilities.

```python
f, axarr = plt.subplots(1, 3, sharex='col', sharey='row', figsize=(15, 5))

for i in range(K):
    Z = post_prob(np.c_[xx.ravel(), yy.ravel()], i)
    Z = Z.reshape(xx.shape)
    axarr[ i ].set_title('k = ' + str(i))
    axarr[ i ].set_xlabel('Attribute #1')
    axarr[ i ].set_ylabel('Attribute #2')
    axarr[ i ].contourf(xx, yy, Z, cmap=plt.cm.hot)
plt.show()
```

## Decision Boundaries or Surfaces of the Bayes' Classifier

We define the Bayes' classifer. This classifier is the one that minimizes the classification error. It simply predicts the class with the highest posterior probability at each input location.

```python
def bayes_classifier(x):
    
    p_c_k_given_x = np.ones((x.shape[ 0 ], K))
    
    for i in range(K):
        p_c_k_given_x[ :, i ] = post_prob(x, i)
    
    return np.argmax(p_c_k_given_x, 1)
```

We plot the decision boundaries of the Bayes' classifer.

```python
plt.figure()
plt.title('Decision Surfaces of the Bayes\' Classifier')
plt.xlabel('Attribute #1')
plt.ylabel('Attribute #2')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
Z = bayes_classifier(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha = 0.4, levels = [-1,0,1,2], colors = (('red','green','blue')))
plt.show()
```

## Minimizing the Expected Cost

We now will consider un-even classification costs. In particular, we will assume the following loss matrix whose i-th row contains the cost of classifiing an instance with true class i, with class j, where j is the number of column.

```python
# We assume class number 2 is the healthy class and class 0 and 1 are the sick classes.
# So we would like to avoid making mistakes in that direction (wrongly classifying as class 2)

# We consider that the cost of classifying a true class 0 instance as class 2 is 100.
# The cost of classifying a true class 1 instance as class 2 is 100 too.

L = np.zeros((K, K))
L[ 0, 1 ] = 1
L[ 0, 2 ] = 100
L[ 1, 0 ] = 1
L[ 1, 2 ] = 100
L[ 2, 0 ] = 1
L[ 2, 1 ] = 1
print(L)
```

### We Define the Minimum Loss Classifier

The minimum loss classifier is the one that outputs the class label $C_j$ minimizing the sum $\sum_k L_{kj}p(C_k|\mathbf{x})$.

```python
def minimum_loss_classifier(x):
    
    p_c_k_given_x = np.ones((x.shape[ 0 ], K))
    
    for i in range(K):
        p_c_k_given_x[ :, i ] = post_prob(x, i)
    
    # We compute the cost of making each prediction by multipling the probabilities by L
    
    weighted_probs = np.dot(p_c_k_given_x, L)
    
    return np.argmin(weighted_probs, 1)
```

We plot the decision boundaries of the minimum loss classifer.

```python
plt.figure()
plt.title('Decision Surfaces of the Minimum Loss Classifier')
plt.xlabel('Attribute #1')
plt.ylabel('Attribute #2')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
Z = minimum_loss_classifier(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha = 0.4, levels = [-1,0,1,2], colors = (('red','green','blue')))
plt.show()
```

We can observe that the region of the input space in which class $\mathcal{C}_2$ is predicted (blue area) is reduced by a lot due to the extra cost of making classification mistakes predicting this class label. 

## Reject Option

We will consider now that we will conly consider decisions with at least 90% of confidence. That is, we only output a decision when the posterior class probability is at least 0.90. For this, we define a new classifier will output a new class label when avoiding decisions.

```python
def reject_classifier(x):
    
    p_c_k_given_x = np.ones((x.shape[ 0 ], K))
    
    for i in range(K):
        p_c_k_given_x[ :, i ] = post_prob(x, i)
        
    # We assign class label K + 1 to the rejected predictions
        
    probs = np.amax(p_c_k_given_x,1)
    predictions = np.argmax(p_c_k_given_x, 1)
    predictions[ probs < 0.90 ] = K
    
    return predictions
```

We plot the decision boundaries of the new classifer.

```python
plt.figure()
plt.title('Decision Surfaces of the Reject Classifier')
plt.xlabel('Attribute #1')
plt.ylabel('Attribute #2')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
Z = reject_classifier(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha = 0.4, levels = [-1,0,1,2,3], colors = (('red','green','blue','gray')))
plt.show()
```

We observe that the reject region (gray area) is located close to the decision boundaries of the Bayes' classifier.

# Practical Exercise

You are working for a company that has developed a new test for a disease. The test is characterized for having a high probability of resulting positive for sick persons. However, sometimes it may fail as indicated by the following table:

|  Test output  | Is the person sick?  | Probability  |
|---|---|---|
|  + | Sick  | 0.9  | 
|  - | Sick  | 0.1  |
|  + |  Healthy |  0.05 | 
|  - | Healthy  | 0.95  |




Let the two class labels be $\mathcal{C}_0 \equiv \text{Healty}$ and $\mathcal{C}_1 \equiv \text{Sick}$ and $x$ be the output of the test, i.e., the observed variables. Therefore the previous table provides all the potential values for the class conditional distributions $p(x|\mathcal{C}_k)$.

## Scenario 1: 

Assume that we have an equal prior probability that a person is either sick or healty, i.e., $p(\mathcal{C}_1)=p(\mathcal{C}_0)=0.5$. 

1. Use Bayes' theorem to compute the fraction of persons that are actually sick from the total number of persons that are positive according to the test. That is, you are asked to compute $p(\mathcal{C}_1|x=+)$.

2. According to the results obtained, is the developed test a very good one? Why?

Ejercicio 1)

$p(sick|+) = \frac{p(+|sick)*p(sick)}{p(+)} = \frac{p(+|sick) p(sick)}{p(+, sick) + p(+, healthy)} = \frac{p(+, sick) p(sick)}{p(+|sick)p(sick) + p(+|healthy) p(healthy)}$

Ahora rellenamos con los datos del enunciado y la tabla

$\frac{0.9 * 0.5}{0.9 * 0.5 + 0.05 * 0.5} = 0.95$

Ejercicio 2)

Parece un buen test, ya que si es positivo, con un 95% de probabilidad la persona realmente est enferma.

## Scenario 2:

Assume now that the disease is very rare and that it only affects to $1\%$ of the popultaion. That is, $p(\mathcal{C}_1)=0.01$. 

1. Use Bayes' theorem to compute the fraction of persons that are actually sick from the total number of persons that are positive according to the test. That is, you are asked to compute $p(\mathcal{C}_1|x=\text{+})$.

2. According to the results obtained, is the developed test a very good one? Why? What is failing in the test in this scenario.

3. Estimate what is the required false positive rate, i.e., $p(x=\text{+}|\mathcal{C}_0)$, of the test so that at least $90\%$ of the persons tested positive are actually sick. 

Assume now that we have $T$ independent tests to diagnose the disease, but that have the same characteristics as those described by the table above. Because the tests are independent, we have that 

$$ p(x_1=\text{+},x_2=\text{+},\ldots, x_T = \text{+}|\mathcal{C}_k)=\prod_{i=1}^T p(x_i=\text{+}|\mathcal{C}_k)$$

In how many of these tests has a person to be positive to so that at least $90\%$ of the persons diagnosed as positive are actually sick. 

Ejercicio 1)

$p(sick|+) = \frac{p(+|sick)*p(sick)}{p(+)} = \frac{p(+|sick) p(sick)}{p(+, sick) + p(+, healthy)} = \frac{p(+, sick) p(sick)}{p(+|sick)p(sick) + p(+|healthy) p(healthy)}$

Ahora rellenamos con los datos del enunciado y la tabla

$\frac{0.9 * 0.01}{0.9 * 0.01 + 0.05 * 0.99} = 0.15$

Ejercicio 2)

Este test es malo, ya que con un test positivo, la probabilidad de estar enfermo es bastante baja. El problema est en los falsos positivos. Para mejorar el test, hay que reducir la probabilidad de los falsos positivos.
