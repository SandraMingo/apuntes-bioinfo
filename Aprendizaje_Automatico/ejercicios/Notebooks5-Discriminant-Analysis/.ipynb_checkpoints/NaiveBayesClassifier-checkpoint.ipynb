# Naive Bayes Classifier Notebook

In this notebook we illustrate the practical use of the naive Bayes classifier. In a first stage, we will show that such a classifier can give un-accurate predictions when the independence assumption is not satisfied in practice. Later on, we will explore its prediciton performance on the Breast Cancer dataset from the UCI repository.

First we load the required packages.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import scipy as sp
```

## Comparison with the Bayes' Classifier

We compare the decision boundaries and class posterior density estimation of the Naive Bayes classifier with the optimal classifier, the Bayes classifier. For that, we will consider the simple example used to illustrate the Bayes classifier in a previous notebook. The class conditional densities, i.e., $p(x|C_k)$ we will be assumed to be Gaussian densities with some pre-defined parameters (chosen at random). The dimensionality of the problem is set to $D=2$ for the sake of visualization. Under the naive Bayes assumption those Gaussian densities will be assumed to factorize. 

```python
# We fix the random seed to guarantee similar results each time the notebook is run

np.random.seed(0)

D = 2
K = 3
means = [ np.random.normal(size = D) for i in range(K) ]
covariances = [ np.random.normal(size = ((D, D))) for i in range(K) ]

# We are careful to guarantee that the covariance matrices are positive-definite

for i in range(K):
    covariances[ i ] = np.dot(covariances[ i ], covariances[ i ].T)
    
# These are the covariance matrices of the naive Bayes classifier. 
# We simply put all entries outside of the diagonal to zero to enforce independence.

covariances_naive = [ covariances[ k ] * np.eye(D) for k in range(K) ]
```

We plot the corresponding Gaussian densities, genrated randomly. The first row shows the actual class conditional densities. The second row shows the corresponding densities obtained under the naive Bayes assumption.

```python
x_min = -2
x_max = 5
y_min = -5
y_max = 5

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

f, axarr = plt.subplots(2, 3, sharex='col', sharey='row', figsize=(15, 10))

for i in range(K):
    mv = stats.multivariate_normal(mean = means[ i ], cov = covariances[ i ])
    Z = mv.pdf(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    axarr[0,i].set_title('k = ' + str(i))
    axarr[0,i].set_xlabel('Attribute #1')
    axarr[0,i].set_ylabel('Attribute #2')
    axarr[0,i].contourf(xx, yy, Z, cmap=plt.cm.hot)

for i in range(K):
    mv = stats.multivariate_normal(mean = means[ i ], cov = covariances_naive[ i ])
    Z = mv.pdf(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    axarr[1,i].set_title('k = ' + str(i))
    axarr[1,i].set_xlabel('Attribute #1')
    axarr[1,i].set_ylabel('Attribute #2')
    axarr[1,i].contourf(xx, yy, Z, cmap=plt.cm.hot)
plt.show()
```

We now specify the class prior probabilities.

```python
class_prior = np.array([ 0.5, 0.2, 0.3])
```

We use the Bayes' rule to compute the class posterior probabilities.

```python
def post_prob(x, k):
    
    p_x_and_ck = np.ones((x.shape[ 0 ], K))
    
    for j in range(K):
        mv = sp.stats.multivariate_normal(mean = means[ j ], cov = covariances[ j ])
        p_x_given_ck = mv.pdf(x)
        p_x_and_ck[ :, j ] = p_x_given_ck * class_prior[ j ]
        
    return p_x_and_ck[ :, k ] / np.sum(p_x_and_ck, axis = 1)
    
def post_prob_naive(x, k):
    
    p_x_and_ck = np.ones((x.shape[ 0 ], K))
    
    for j in range(K):
        mv = sp.stats.multivariate_normal(mean = means[ j ], cov = covariances_naive[ j ])
        p_x_given_ck = mv.pdf(x)
        p_x_and_ck[ :, j ] = p_x_given_ck * class_prior[ j ]
        
    return p_x_and_ck[ :, k ] / np.sum(p_x_and_ck, axis = 1)
    
```

We plot the class posterior probabilities. The first rows show the actual class posterior probabilities. The second row shows the ones obtained by the Naive Bayes classifier.

```python
f, axarr = plt.subplots(2, 3, sharex='col', sharey='row', figsize=(15, 10))

for i in range(K):
    Z = post_prob(np.c_[xx.ravel(), yy.ravel()], i)
    Z = Z.reshape(xx.shape)
    axarr[ 0, i ].set_title('k = ' + str(i))
    axarr[ 0, i ].set_xlabel('Attribute #1')
    axarr[ 0, i ].set_ylabel('Attribute #2')
    axarr[ 0, i ].contourf(xx, yy, Z, cmap=plt.cm.hot)

for i in range(K):
    Z = post_prob_naive(np.c_[xx.ravel(), yy.ravel()], i)
    Z = Z.reshape(xx.shape)
    axarr[ 1, i ].set_title('k = ' + str(i))
    axarr[ 1, i ].set_xlabel('Attribute #1')
    axarr[ 1, i ].set_ylabel('Attribute #2')
    axarr[ 1, i ].contourf(xx, yy, Z, cmap=plt.cm.hot)
    
plt.show()
```

We now define each classifier. The Bayes' classifier and the Naive Bayes classifier.

```python
def bayes_classifier(x):
    
    p_c_k_given_x = np.ones((x.shape[ 0 ], K))
    
    for i in range(K):
        p_c_k_given_x[ :, i ] = post_prob(x, i)
    
    return np.argmax(p_c_k_given_x, 1)

def naive_bayes_classifier(x):
    
    p_c_k_given_x = np.ones((x.shape[ 0 ], K))
    
    for i in range(K):
        p_c_k_given_x[ :, i ] = post_prob_naive(x, i)
    
    return np.argmax(p_c_k_given_x, 1)
```

We plot the corresponding decision boundaries or surfaces.

```python
f, axarr = plt.subplots(1, 2, sharex='col', sharey='row', figsize=(15, 5))

axarr[ 0 ].set_title('Decision Surfaces of the Bayes\' Classifier')
axarr[ 0 ].set_xlabel('Attribute #1')
axarr[ 0 ].set_ylabel('Attribute #2')
Z = bayes_classifier(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
axarr[ 0 ].contourf(xx, yy, Z, alpha = 0.4, levels = [-1,0,1,2], colors = (('red','green','blue')))

axarr[ 1 ].set_title('Decision Surfaces of the Naive Bayes Classifier')
axarr[ 1 ].set_xlabel('Attribute #1')
axarr[ 1 ].set_ylabel('Attribute #2')
Z = naive_bayes_classifier(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
axarr[ 1 ].contourf(xx, yy, Z, alpha = 0.4, levels = [-1,0,1,2], colors = (('red','green','blue')))

plt.show()
```

We can observe that the decision boundaries or surfaces of the Naive Bayes Classifier are quite different from the ones of the optimal classifier, the Bayes' classifier. This indicates that the Naive Bayes Classifier will lead to an increased prediction error. Therefore, when the independence assumption is not satisfied in practice, the Naive Bayes classifier can give bad prediction results, compared to the Bayes' classifier.

## Evaluation of the Naive Bayes Classifier on an Actual Dataset

In spite of the limitations described before, the Naive Bayes Classifier is still a popular first option to consider when addressing a classification problem. Its main advantages are its simplicity and robustness to over-fitting, as a consequence of the lack of too many parameters. In this part of the notebook, we will consider the implementation provided in Scikit-learn, which is limited to Gaussian, Bernoulli or Multinulli distributions. Since the dataset considered conatins only real variables, we will focus on the Gaussian case. 

We will estimate the prediction performance of the classifier on the dataset considered using a random partition of the data into train and test sets.

First, we import the required packages.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Loading the Data

We load the data from a CSV file using pandas. The dataset considered is 'Wisconsin Diagnostic Breast Cancer' from the UCI repository. We ignore the first attribute in this dataset, which corresponds to the ID of each sample. The task of interest is to predict whether or not a cell corresponds to a malignant tumor. In this dataset all attributes take real values. These include area, smoothness, perimeter, etc. The PDF assumed by the Naive Bayes Classifier will hence be Gaussian, with mean $\mu_{jk}$ and variance $\sigma_{jk}^2$ estimated by maximum likelihood.

$$p(x_j|\mathcal{C}_k) = \frac{1}{\sqrt{2\pi \sigma_{jk}^2}} \exp \left\{-\frac{(x_j-\mu_{jk})^2}{2\sigma_{jk}^2} \right\}$$

```python
data = pd.read_csv('wdbc.csv')
X = data.values[ :, 2:].astype(float)
y = (data.values[ :, 1 ] == 'B').astype(int)
```

### Data Partition into Train and Test


We randomly split the data into train and test partitions. We use 70% of the data for training and 30% for testing. 


```python
X_train, X_test, y_train, y_test = train_test_split( \
    X, y, test_size=0.3, random_state=1)
```

### Data Standardization

We standardize the data to have zero mean and unit standard deviation.

```python
scaler = preprocessing.StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

We check the standardization process.

```python
print(np.var(X_train[:,0]))
print(np.var(X_train[:,1]))
print(np.var(X_train_scaled[:,0]))
print(np.var(X_train_scaled[:,1]))
```

### Evaluating the Performance on the Test Set

We train the classifier.

```python
nb = GaussianNB()
nb.fit(X_train_scaled, y_train)
```

We compute the confusion matrix on the test set.

```python
y_pred = nb.predict(X_test_scaled)

conf = confusion_matrix(y_test, y_pred)

# The matrix conf contains:
# TN | FN
# FP | TP

TN = conf[0][0]
TP = conf[1][1]
FP = conf[0][1]
FN = conf[1][0]

print(conf)
print('Predicion accuracy is: %f' % ((TP + TN) / (TN + TP + FP + FN)))
print('True postive rate is: %f' % (TP / (TP + FN)))
print('True negative rate is: %f\n' % (TN / (TN + FP)))
```

Eventhough these results look very good they can be easily outperformed by other classification systems that we will also cover soon.

### Classification or Decision Frontier Visualization

We show how the decision boundaries of the classifier look like for different values of the hyper-parameters. For simplicity, we only consider the first two attributes of the previous problem. We use all the available data.

```python
X = X[:, [0, 1]]
```

As previously, the first step is to standardize the data. 

```python
X_scaled = preprocessing.scale(X)
```

We check that the standardization process has been successful.

```python
print(np.var(X[:,0]))
print(np.var(X[:,1]))
print(np.var(X_scaled[:,0]))
print(np.var(X_scaled[:,1]))
```

We train the classifier.

```python
nb.fit(X_scaled, y)
```

We plot the resulting decision boundaries.

```python
x_min, x_max = X_scaled[:, 0].min() - .5, X_scaled[:, 0].max() + .5
y_min, y_max = X_scaled[:, 1].min() - .5, X_scaled[:, 1].max() + .5

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

Z = nb.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure()
plt.contourf(xx, yy, Z, alpha=0.4, cmap=plt.cm.Set1)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, s=20, edgecolor='k',\
        cmap=plt.cm.Set1)
plt.title('Decision Boundaries of the Naive Bayes Classifier')
plt.xlabel('Attr. 1')
plt.ylabel('Attr. 2')

plt.show()
```
