## Cálculos sobre grafos: NetworkX

```python
import networkx as nx

G = nx.Graph()
G.add_nodes_from([0, 1, 2, 3, 4, 5, 6, 7])
G.add_edges_from([(0, 1), (0, 2), (1, 3), (2, 4), (3, 5), (4, 6), (5, 7), (6, 7), (1, 4), (3, 4), (5, 6)])

print(G.nodes())
print(G.edges())
```

```python
import matplotlib.pyplot as plt
import networkx as nx

#probad con varios órdenes
G = nx.gnp_random_graph(1000, 0.02)
L=nx.degree_histogram(G)

plt.bar(range(len(L)),L, width=0.80, color='b')
plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
plt.show()
```

La distribución que debe salir es de Poisson, ya que se ha generado aleatoriamente. Este mismo código con una red biológica debería dar una distribución libre de escala (distribución de potencias); no debe salir una Poisson. La Poisson sale cuando las ramas tienen la misma probabilidad de salir. Esto no ocurre en un grafo biológico, ya que la probabilidad de que dos nodos conectados a un tercero estén conectados también entre sí es mucho más elevada.
